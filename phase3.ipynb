## Advance operation with buiness examples 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("===== Zomato soles analysis =====")
print("\n Sales Data Shape",sales_data.shape)
print("\n SAmple Data of First three restaurant:\n ",sales_data[0:3])
print("\n SAmple Data of First three restaurant:\n ",sales_data[:,2:])
```

```python
print(" Total sales per year :",np.sum(sales_data[:,1:], axis=0))

print("\n Minimum sales per restaurant :", np.min(sales_data[:,1:], axis=0))

print("\n Max sales per restaurant :", np.max(sales_data[:,1:], axis=0))

print("\n Total sales per restaurant :", np.sum(sales_data[:,1:], axis=1))

print("\n Mean sales per restaurant :", np.mean(sales_data[:,1:], axis=1))
```

```python
cumsum = np.cumsum(sales_data[:,1:],axis=1)

print(cumsum)

plt.figure(figsize=(10,6))
plt.plot(np.mean( cumsum , axis=0))
plt.title("Average cumulative sales accross all restaurant ")
plt.xlabel("years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()

```
